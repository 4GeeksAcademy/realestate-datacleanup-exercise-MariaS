Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
valdemorillo = ds[ds["level5"]== "Valdemorillo"]
valdemorillo_price = valdemorillo["price"]
media_valdemorillo = valdemorillo_price.mean()
galapagar = ds[ds["level5"]=="Galapagar"]
galapagar_price = galapagar["price"]
media_galapagar = galapagar_price.mean()
print(f"Promedio de precios de Valdemorillo {media_valdemorillo}")
print(f"Promedio de precios de Galapagar {media_galapagar}")

#Ambos promedios son muy similares,Esto indica que las propiedades en ambas localidades tienen precios promedio comparables pero 
#Pero no me parece informacion suficiente para sacar conclusiones por eso saco la mediana y desviacion

desviacion_valdemorillo = valdemorillo_price.std()
desviacion_galapagar = galapagar_price.std()
print("Desviación estándar de Valdemorillo:", desviacion_valdemorillo)
print("Desviación estándar de Galapagar:", desviacion_galapagar)

mediana_valdemorillo = valdemorillo_price.median()
mediana_galapagar = galapagar_price.median()
print("Mediana de Valdemorillo:", mediana_valdemorillo)
print("Mediana de Galapagar:", mediana_galapagar)

# Los precios en ambas localidades tienen alta variación (>=50% de la media).
# En Galapagar, los precios son más equilibrados (mediana cercana al promedio).
# Conclusión: Los precios son similares en promedio, pero con gran variación: indica que hay una mezcla de propiedades económicas y de lujo en ambas localidades
# Galapagar tiene precios más equilibrados que Valdemorillo







```

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
poblaciones_filtradas = ["Fuenlabrada", "Leganes", "Getafe", "Alcorcon"]
subconjunto = ds[ds["level5"].isin(poblaciones_filtradas)]
subconjunto_precios = subconjunto["price"]
subconjunto_sin_nulos = subconjunto[subconjunto["price"].notnull()]
agrupacion_subconjunto = subconjunto.groupby("level5")["price"]
subconjunto_sin_nulos["price_normalized"] = subconjunto.groupby("level5")["price"].transform(lambda x: (x - x.min()) / (x.max() - x.min()))


import matplotlib.pyplot as plt

for poblacion in poblaciones_filtradas:
    precios = subconjunto_sin_nulos[subconjunto_sin_nulos["level5"] == poblacion]["price_normalized"]
    
    plt.hist(precios, bins=10, alpha=0.5, label=poblacion) 

# Configuración del gráfico
plt.legend()  
plt.title("Distribución normalizada de precios por población")  
plt.xlabel("Precio Normalizado")  
plt.ylabel("Frecuencia")  
plt.show()  
```

Fuenlabrada tiene precios más concentrados hacia los valores bajos.La mayoría de las casas en Fuenlabrada tienen precios relativamente bajos en comparación con el rango de precios normalizado de esa población.

Getafe muestra una mayor dispersión, indicando una mayor diversidad de precios. Hay casas tanto en rangos bajos como en rangos más altos, lo que indica mayor variedad en los precios dentro de esta población.

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
#"Valdemoro", "Chinchon", "El Escorial", "Las Rozas de Madrid"
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')
poblaciones = ["Valdemoro", "Los Molinos", "El Escorial", "Las Rozas de Madrid"]
nuevo_conjunto = ds[ds["level5"].isin(poblaciones)]

nuevo_conjunto = nuevo_conjunto[nuevo_conjunto["surface"].notnull()]  # Eliminar nulos
nuevo_conjunto = nuevo_conjunto[nuevo_conjunto["surface"] > 0]        # Eliminar ceros

nuevo_conjunto["pps"] = nuevo_conjunto["price"] / nuevo_conjunto["surface"]

print(nuevo_conjunto[["price", "level5", "surface", "pps"]].head())

import matplotlib.pyplot as plt

agrupado_por_poblacion = nuevo_conjunto.groupby("level5")


fig, axs = plt.subplots(4, 1, figsize=(8, 16), sharex=True)

for (nombre_poblacion, datos), ax in zip(agrupado_por_poblacion, axs):
    ax.scatter(datos["surface"], datos["price"], alpha=0.7, label=nombre_poblacion, color="blue")
    ax.set_title(f"Precio vs. Superficie ({nombre_poblacion})")
    ax.set_xlabel("Superficie (m²)")
    ax.set_ylabel("Precio (€)")
    ax.legend()

plt.tight_layout()
plt.show()

```
